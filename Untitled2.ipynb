```python
#Importando a biblioteca
import pandas as pd
```

```python
#Leitura dos arquivos
df1 = pd.read_excel("/content/drive/MyDrive/DataBases/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/DataBases/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/DataBases/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/DataBases/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/DataBases/Salvador.xlsx")
```

```python
#juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#exibindo as 5 primeiras linhas
df.head()
```

```python
#exibindo as 5 ultimas linhas
df.tail()
```

```python
#verificar o tipo de dado de cada coluna
df.dtypes
```

```python
#alterando o tipo de dado da coluna lojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

```python
#consultando linhas com valores faltantes
df.isnull().sum()
```

```python
df["Vendas"].fillna(0, inplace=True)
```

```python
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df["Data"] = pd.to_datetime(df["Data"])
```
